```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
data = pd.read_csv('netflix_titles.csv', encoding='latin-1')
print(data)
```

```python
data.describe
```

```python
data.columns
```

```python
data.dtypes
```

```python
dt=pd.DataFrame(data)
dt
```

```python
data = dt.iloc[ :,:10]
data
```

```python
data.isnull().sum()
```

```python
df = data.dropna(subset=['type','rating', 'country','duration','release_year'])
df
```

```python
df.isnull().sum()
```

```python
df.type
```

```python
type_count = df['type'].value_counts ()
print(type_count)
```

```python
plt.figure(figsize=(6,4))
plt.barh(type_count.index , type_count.values, color= ['blue' , 'green'])
plt.title ('TV_SHOWS VS MOVIES')
plt.xlabel('Types')
plt.ylabel('counts')
plt.tight_layout()
plt.show
```

```python
rating_count = df['rating'].value_counts ()
rating_count
```

```python
plt.figure(figsize=(18,100))
plt.pie ( rating_count ,labels =  rating_count.index ,autopct='%1.1f%%' , startangle=90)
plt.title('RATING OF TV_SHOWS AND MOVIES')
plt.show()
```

```python
country_count= df['country'].value_counts().head (10)
country_count
```

```python
plt.figure(figsize=(10,6))
plt.scatter(country_count.index , country_count.values,color='black' )
plt.xlabel('NAME OF THE COUNTRY')
plt.ylabel('NUMBER OF MOVIES')
plt.title('PRODUCTION OF MOVIES ACCORDING TO COUNTRY')

plt.grid()
plt.tight_layout()
plt.show()
```

```python
movie_df = df[df['type']=='Movie'].copy()
movie_df['duration_int']= movie_df['duration'].str.replace( 'min','').str.strip().astype(int)

plt.figure(figsize=(8,8))
plt.hist(movie_df['duration_int'], bins=30, color='cyan', edgecolor='black')
plt.title('DISTRIBUTION OF MOVIE DURATIONS')
plt.xlabel('Number of Movies')
plt.ylabel('DURATION(MINUTES)')
plt.show()
```
